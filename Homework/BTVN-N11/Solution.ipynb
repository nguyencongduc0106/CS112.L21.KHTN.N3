Nếu không có mảng nào thỏa thì in ra -1.

# **2. Decomposition**
- Sắp xếp mảng giảm dần.
- Lấy 3 phần tử liên tiếp và kiểm tra điều kiện (${a<b+c}$).
- Chọn mảng con có 3 phần tử thỏa đề. 

# **3. Pattern Recognition**
Sort mảng giảm dần rồi kiểm tra điều kiện a[i]<a[i+1]+a[i+2] để tối ưu việc kiểm tra

# **4. Algorithm**

```python
def function(sticks):
  sticks.sort(reverse = True)
  for i in range(len(sticks)-2):
    if(sticks[i] < sticks[i+1] + sticks[i+2]):
      return [sticks[i], sticks[i+1], sticks[i+2]]
  return -1
N = int(input())
sticks = list(map(int, input().split()))
print(function(sticks))
```

# **5.Complexity:** O(nlogn)
