# **1.Chương trình:**

```python
def findmax(a):
    n = len(a)
    if n == 1:
        return 0
    if n == 2:
      if a[0] > a[1]:
        return 0
      else:
        return 1

    mid = len(a) // 2
    left = a[0:mid]
    right = a[mid:len(a)]
    L = findmax(left)
    R = findmax(right) + mid
    if a[L] > a[R]:
      return L
    else:
      return R

a = [int(x) for x in input().split()]
print(a)
print("Vị trí phần tử lớn nhất: "+str(findmax(a)))
```

# **2.Recurrence relation:**
$T(n) = T(n/2) + n$

$O(n) = nlog(n)$
